# Парсинг заголовков публикаций новостного архива ресурса SpectrumNews (https://www.spectrumnews.org/news-archive/) за интересуемый год

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

Следующие несколько блоков представляют собой пример того, каким именно образом будет производиться обращение к необходимому HTML-коду. Пример производится на последней на данный момент публикации в архиве.

```python
response = requests.get('https://www.spectrumnews.org/news-archive/page/1/')
soup=BeautifulSoup(response.text, "lxml")
soup
```

```python
link = soup.find_all(attrs = {"class":"article-content-col"})[0].find("h2").find('a').get("href")
link
```

```python
title = soup.find_all(attrs = {"class":"article-content-col"})[0].find('h2').get_text()
title
```

```python
author = soup.find_all(attrs = {"class":"article-content-col"})[0].find(attrs = {"class":"author-header-authors"}).find('a').get_text()
author
```

```python
date = soup.find_all(attrs = {"class":"article-content-col"})[0].find(attrs = {"class":"author-header-date"}).get_text()
date
```

```python
abstract=soup.find_all(attrs = {"class":"article-content-col"})[0].find('p').get_text()
abstract
```

```python
comments = soup.find_all(attrs = {"class":"article-content-col"})[0].find(attrs = {"class":"article-comments"}).get_text()
comments
```

```python
print(link)
print(title)
print(author)
print(date)
print(abstract)
print(comments)
```

На момент 7 февраля 2019 года, публикации за 2018 год в архиве начинаются со страницы 3, и заканчиваются на странице. Чтобы выгрузить данные за 2018 год, необходимо указать эти страницы.

```python
year=input('Введите год, публикации за который необходимо выгрузить: ')
```

```python
page1=int(input("Введите номер страницы новостного архива SpectrumNews, с которого начинаются статьи необходимого года: "))
```

```python
page2=int(input('Введите номер страницы новостного архива SpectrumNews, на которой заканичваются статьи необходимого года: '))
```

```python
p=1
for n in range(page1, page2+1):
    response = requests.get('https://www.spectrumnews.org/news-archive/page/{}/'.format(n))
    soup=BeautifulSoup(response.text, "lxml")
    for i in range (10):
        df.loc[p, link_col] = soup.find_all(attrs = {"class":"article-content-col"})[i].find("h2").find('a').get("href")
        df.loc[p, title_col] = soup.find_all(attrs = {"class":"article-content-col"})[i].find('h2').get_text()
        df.loc[p, author_col] = soup.find_all(attrs = {"class":"article-content-col"})[i].find(attrs = {"class":"author-header-authors"}).find('a').get_text()
        df.loc[p, date_col] = soup.find_all(attrs = {"class":"article-content-col"})[i].find(attrs = {"class":"author-header-date"}).get_text()
        df.loc[p, abstract_col] = soup.find_all(attrs = {"class":"article-content-col"})[i].find('p').get_text()
        df.loc[p, comments_col] = soup.find_all(attrs = {"class":"article-content-col"})[i].find(attrs = {"class":"article-comments"}).get_text()
        p+=1
```

```python
df
```

Следующая команда очищает полученный массив данных от статей за годы, которые нас не интересуют

```python
df = df[df[date_col].str.contains(year)]
df
```

Сохранение данных в формате .xlsx.

```python
df.to_excel('SpectrumNews_Publications for {}.xlsx'.format(year))
```
